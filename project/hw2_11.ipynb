```python
import numpy
from sympy import*
import matplotlib
import cmath
from sympy.plotting import plot
init_printing()
```

```python
x = Symbol('x')
f=-5*x**2+10*x+150
f
```

```python
fun=plot(f,(x, -2, 2), show=False)  #Построили график
fun.show()
```

```python
f1=diff(f) #Нашли производную для нахождения вершины
f1
```

```python
fun=plot(f1,(x, -1.2, 0.8), show=False)  
fun.show()
```

```python
print(f'Вершина равна', solve (f1) ) # Находим вершину
```

```python
x1,x2=solve(f)  #Находим корни функции
print(f'Первый корень = ', {round(x1.evalf(), 4)}, 'Второй корень = ', {round(x2.evalf(), 4)})
```

```python
extremum = solve(f1)
print(f'Функция убывает от -∞  до {extremum } и Функция возрастает с {extremum } до +∞')
```

```python
print(f'Функция {f} меньше нуля от -∞ до {round(x1.evalf(),4)} и от {round(x2.evalf(),4)} до +∞')
print(f'Функция {f} больше нуля от {round(x1.evalf(),4)} и до {round(x2.evalf(),4)}')
```

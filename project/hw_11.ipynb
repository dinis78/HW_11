```python
import numpy
from sympy import*
import matplotlib
import cmath
from sympy.plotting import plot
init_printing()
```

```python
x = Symbol('x')
f= -12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30
```

```python
fun=plot(f, (x, -1000, 1000), show=False)  #Построить график
fun.show()
```

```python
f1=diff(f)
```

```python
fun=plot(f1, show=False)  #Находим производную
fun.show()
```
